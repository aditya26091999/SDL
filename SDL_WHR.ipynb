```python
import pandas as pd
import matplotlib.pyplot as plt

mydata = pd.read_csv('2015.csv')
mydata = pd.DataFrame(mydata)
mydata = mydata.rename({'Country':'C','Region':'R','Happiness Rank':'Hr','Happiness Score':'Hs',
                        'Standard Error':'Se','Economy (GDP per Capita)':'Ec',
                        'Family':'Fa','Health (Life Expectancy)':'Hl','Freedom':'Fr',
                        'Trust (Government Corruption)':'Tr',
                       'Generosity':'Gn','Dystopia Residual':'DR'},
                    axis=1)


```

```python
%matplotlib inline 
x = mydata['Hr']
y = mydata['Hs']
plt.title('HR vs HS')
plt.xlabel('Happiness Rank')
plt.ylabel('Happiness Score')
plt.plot(x,y)
plt.show()
```

```python
%matplotlib inline
plt.title('Generosity vs Family')
x = mydata['Gn']
y = mydata['Fa']
plt.xlabel('Generosity')
plt.ylabel('Family')
plt.scatter(x,y)
```

```python
%matplotlib inline
plt.title('Trust on Govt vs Freedom')
x = mydata['Tr']
y = mydata['Fr']
plt.xlabel('Trust on Govt.')
plt.ylabel('Freedom')
plt.scatter(x,y)
```
